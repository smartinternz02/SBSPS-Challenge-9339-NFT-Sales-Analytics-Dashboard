```python
df['Active_Market_Wallets_cumsum'] = df['Active_Market_Wallets_cumsum'].fillna(df['Active_Market_Wallets_cumsum'].mean())
df['Secondary_Sales_cumsum'] = df['Secondary_Sales_cumsum'].fillna(df['Secondary_Sales_cumsum'].mean())
df['AverageUSD_cum '] = df['AverageUSD_cum'].fillna(df['AverageUSD_cum'].mean())
df['Sales_USD'] = df['Sales_USD'].fillna(df['Sales_USD'].mean())
df['Number_of_Sales'] = df['Number_of_Sales'].fillna(df['Number_of_Sales'].mean())
df['Active_Market_Wallets'] = df['Active_Market_Wallets'].fillna(df['Active_Market_Wallets'].mean())
df['Primary_Sales'] = df['Primary_Sales'].fillna(df['Primary_Sales'].mean())
```

```python
import matplotlib.pyplot as plt

# Draw Plot
def plot_df(df, x, y, title="", xlabel='Date', ylabel='NFT', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df, x=df.index, y=df.Number_of_Sales, title='Daily NFT Sales.') 
```

```python
plot_df(df, x=df.index, y=df.AverageUSD_cum, title='Average Sale Price.') 
```

```python
plot_df(df, x=df.index, y=df.Active_Market_Wallets_cumsum, title='Active Market Wallets')  
```

```python
plot_df(df, x=df.index, y=df.Sales_USD_cumsum, title='Active Sales')
```
